```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV file into a DataFrame
csv_file_path = '../XRPUSDT_Dataset/XRP_dataset.csv'  # Replace with your CSV file path
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df = df.drop(columns=['Unnamed: 0'])
df = df[0:500]
print(df.head())
```

```python
# Calculate 10 Min Moving Average
sma2 = df['close'].rolling(window=2).mean().shift(-1)

# Calculate 30 Min Moving Average
sma6 = df['close'].rolling(window=6).mean()
```

```python
# Plotting
plt.figure(figsize = (10, 6))
plt.plot(df["close"], label = 'Original')
plt.plot(sma2, label = 'MA', color = 'red')
plt.title('20 Min Moving Average (MA)')
plt.legend()
plt.show()
```

```python
# Plotting
plt.figure(figsize = (10, 6))
plt.plot(df["close"], label = 'Original')
plt.plot(sma6, label = 'SMA', color = 'red')
plt.title('30 Min Moving Average (SMA)')
plt.legend()
plt.show()
```
